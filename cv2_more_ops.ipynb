```
#to take input from camera
cap=cv2.VideoCapture(0)#--to start camera--first camera
#print(cap.isOpened())#to check if camera is open/started
#camera started or not
if cap.isOpened():
    print("camera is  ready")
else:
    print("camera not started")
#now can take read input from camera
status,img=cap.read()#take first picture
status1,img1=cap.read()#take 2nd picture
cv2_imshow(img)
#to close camera
cap.release()
#to close window when 'q' is pressed
if cv2.waitKey(10) 7 0xff == ord('q')
  break
 #we can also draw patterns in the image hence will put effect on camera
cv2.line(img,(0,0),(200,300),(0,225,0),3)#same way with rectangle and othersx
```

```
#adding plug-in
plugin=cv2.VideoWriter_fourcc(*'XVID')
#saving video
output=cv2.VideoWriter('class.mp4',plugin,20,(640,480))#file to store,plugin name,fps,camera resolution(width,height)
#give file type .avi if .mp4 not supported
#can draw any pattern just before saving video
output=cv2.VideoWriter('class.mp4',plugin,80,(640,480))#80 fps will make it fastforward
output=cv2.VideoWriter('class.mp4',plugin,3,(640,480))#3 fps makes it slowmo
```
