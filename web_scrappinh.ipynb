```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url="https://en.wikipedia.org/wiki/Cicada"
```

```
#sending request to url and taking the response
response=requests.get(url)
response
```

```
response.text
```

```
#parsing liberary - like - html.parser and lxml etc..
soup=BeautifulSoup(response.text, "lxml" )#will contain the html coscript of the page
```

```
#to search and scrap data from soup.text
#for Alink in soup.find_all('a'):
  #print(Alink)
 # print(Alink.get('href'))
```

```
for ptag in soup.find_all('p'):
    #print(ptag)
    #now second loop for ptag
    for btag in ptag.find_all('b'):
      print(btag.text) #printing data under b-tags inside p-tags simply the content   
```
